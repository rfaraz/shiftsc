# 🏋 Training the Model

Great! Now that we have processed the data, we are finally ready to start training the model. Below, we have already implemented a classic model that you can use, `random forest`. However, it is important to note that random forest classifiers fall under decision trees, which makes it difficult to implement fairness and safety improvement techniques. On top of that, they typically perform worse than their neural network counterparts.

So, after you try out the random forest classifier, implement a neural network model. We have provided you the `sklearn` library, which contains a lot of strong classifiers. Use the [documentation](https://scikit-learn.org/stable/) to help guide you more in defining and using the trained model.

```python
# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# Train model (Random Forest)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Evaluate
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# # Write a neural network
from sklearn.neural_network import MLPClassifier # Here is one example
```
